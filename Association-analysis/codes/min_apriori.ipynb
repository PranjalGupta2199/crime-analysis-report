```python
freq_itemsets = []
freq_itemsets.append([[i] for i in range(1,11)])
freq_itemsets
```

```python
sup = {}
for i in range(1,11):
    sup[tuple([i])]=1
sup
```

```python
def has_infrequent_subset(kitemset,kplusone_item, k):
    kitems=list(itertools.combinations(kplusone_item, k+1))
    kitems=[list(a) for a in kitems]
    for kitem in kitems:
        if kitem not in kitemset:
            return 1
    return 0
```

```python
def apriori_gen(kitemset,k):
    pos_kplusone_itemset = []
    for i in range(len(kitemset)):
        for j in range(i+1,len(kitemset)):
            kplusone_item = list(set(kitemset[i]).union(set(kitemset[j])))
            if(len(kplusone_item)==k+2):
                flag=has_infrequent_subset(kitemset,kplusone_item, k)
                if(flag==0):
                    pos_kplusone_itemset.append(kplusone_item)
    
    pos_kplusone_itemset = set(tuple(sorted(sub)) for sub in pos_kplusone_itemset)
    pos_kplusone_itemset = [list(i) for i in pos_kplusone_itemset]
    return pos_kplusone_itemset
```

```python
def apriori(min_sup):
    k=0
    while(1):
        kplusone_itemset = []
        kitemset = freq_itemsets[k]
        pos_kplusone_itemset = apriori_gen(kitemset,k)
        for kplusone_item in pos_kplusone_itemset:
            sumi=0
            for x in range(len(crimes)):
                mini=10000
                for y in kplusone_item:
                    mini=min(mini,crimes.iloc[x][y])
                sumi+=mini
            if sumi>=min_sup:
                sup[tuple(kplusone_item)]=sumi
                kplusone_itemset.append(kplusone_item)
        if(len(kplusone_itemset)==0):
           break
        #print(kplusone_itemset)
        freq_itemsets.append(kplusone_itemset)
        k=k+1
    return(freq_itemsets)
```

```python
min_sup=0.35
freq_itemsets = apriori(min_sup)

crime = ["", "Murder", "Rape", "Foeticide", "Kidnapping and Abduction", "Abetment of suicide", "Exposure and abandonment", "Procuration of minor girls", "Selling of girls for prostitution", "Prohibition of child marriage act", "Other Crimes"]
for k_itemsets in freq_itemsets:
    for itemset in k_itemsets:
        print("[", end =" ")
        for item in itemset:
            print(crime[item]+", ", end =" ")
        print("]")
```

```python
def get_rules(min_conf):
    rules = []
    for i in range(1,len(freq_itemsets)):
        for itemset in freq_itemsets[i]:
            #print(itemset)
            for j in range(1, len(itemset)):
                listx=list(itertools.combinations(itemset, j))
                for x in listx:
                    conf = sup[tuple(itemset)]/sup[tuple(x)]
                    x=set(x)
                    y=set(itemset).difference(x)
                    if conf>=min_conf:
                        print("conf of",tuple([x,y]),"=",conf)
                        rules.append(tuple([x,y]))
    return rules
```

```python
min_conf=0.5
rules = get_rules(min_conf)
```
